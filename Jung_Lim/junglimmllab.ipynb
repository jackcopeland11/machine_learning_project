```python
# converting these columns into floats by taking the dollar sign out

orders[['Sales','Profit']] = orders[['Sales','Profit']].replace('[\$,]', '', regex=True).astype(float)
```

## Why did customers return orders?

```python
# renaming column so i can merge them by the Order ID column

orders.rename(columns = {'Order.ID':'Order ID'}, inplace = True)
```

```python
# merging

returndesc = pd.merge(orders, returns , on = 'Order ID', how = 'inner')
```

```python
returndesc
```

```python
# checking that none of the rows are repeated

duplicate = returndesc[returndesc.duplicated()]
duplicate
```

```python
returndesc['Year'] = returndesc['Order.Date'].str[-2:]
returndesc['Year'] = '20' + returndesc['Year'].astype(str) 
returndesc['Year']
```

```python
# Total Profit lost each year

agg_func_describe = {'Profit': ['sum']}

returndesc.groupby(['Year']).agg(agg_func_describe)
```

```python
returnregion = returndesc['Region_y'].value_counts()
returnregion = pd.DataFrame(returnregion)
returnregion = returnregion.reset_index()
```

```python
orderregion = orders['Region'].value_counts()
orderregion = pd.DataFrame(orderregion)
orderregion = orderregion.reset_index()
```

```python
merged = pd.merge(returnregion, orderregion , on = 'index', how = 'inner')
merged
```

```python
merged['ratio'] = merged['Region_y']/merged['Region']
ratio = merged.sort_values(['ratio'], ascending=[False])
ratio
```

```python
returncat = returndesc['Sub.Category'].value_counts()
returncat = pd.DataFrame(returncat)
returncat = returncat.reset_index()

ordercat = orders['Sub.Category'].value_counts()
ordercat = pd.DataFrame(ordercat)
ordercat = ordercat.reset_index()
```
