```python
import numpy as np
import pandas as pd
```

```python
orders = pd.read_csv('data/orders.csv', index_col=0)
returns = pd.read_csv('data/returns.csv', index_col=0)
```

```python
returns = returns.reset_index()
returns
```

# Question 3
## Part 1

```python
orders['Profit'] = orders.Profit.apply(lambda x: x.replace('$','')).apply(lambda x: x.replace(',','')).astype(float)
```

```python
orders['Sales'] = orders.Sales.apply(lambda x: x.replace('$','')).apply(lambda x: x.replace(',','')).astype(float)
```

```python
orders
```

```python
combined = pd.merge(orders, returns, how = 'left', left_on='Order.ID', right_on = 'Order ID')
combined
```

```python
combined[~combined.Returned.isnull()]['Profit'].sum()
```

## Part 2

```python
return_vals = combined[~combined.Returned.isnull()]
customer_returns = return_vals['Customer.ID'].value_counts()
customer_returns[customer_returns > 1]
```

```python
customer_returns[customer_returns > 5]
```

## Part 3

```python
region_returns = combined.groupby('Region_x')['Order.ID', 'Returned'].agg('count')
region_returns['ratio'] = (region_returns['Returned'] / region_returns['Order.ID'])
region_returns.sort_values('ratio', ascending = False)
```

## Part 4

```python
cat_returns = combined.groupby('Category')['Order.ID', 'Returned'].agg('count')
cat_returns['ratio'] = (cat_returns['Returned'] / cat_returns['Order.ID'])
cat_returns.sort_values('ratio', ascending = False)
```
